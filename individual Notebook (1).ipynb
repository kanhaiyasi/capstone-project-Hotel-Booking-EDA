##3.DATA WRANGLING

```python
# Making a new column 'kids' joining two columns i.e 'babies' and 'children'.
df['kids']=df['babies'] + df['children']
```

```python
# Making a new column 'total stays' by adding two columns 'stays_in_week_nights' and 'stays_in_weekend_nights'.
df['total_stays']=df['stays_in_week_nights']+df['stays_in_weekend_nights']
```

```python
# Making a new column actual guest based on no cancellation
actual_guests = df.loc[df["is_canceled"] == 0]
# Now creating a new column revenue by multiply adr with total stays
actual_guests['revenue'] = (actual_guests['adr'] * (actual_guests['total_stays']))/1000000
```

```python
df['guest']= df['adults']+df['children']+df['babies']
df['guest'] = df.guest.map( lambda x: 'Single' if x == 1 else 'Family')
df['kids_count'] = df.kids.map( lambda x: 'no_kids' if x == 0 else 'kids')
```

3. Agent most number of booking

```python
d1 = pd.DataFrame(df['agent'].value_counts()).reset_index().rename(columns = {'index':'agent','agent':'num_of_bookings'}).sort_values(by = 'num_of_bookings', ascending = False)
d1.drop(d1[d1['agent'] == 0].index, inplace = True)            # 0 represents that booking is not made by an agent
d1 = d1[:10]                                                   # Selecting top 10 performing agents
plt.figure(figsize = (7,2))
sns.barplot(x = 'agent', y = 'num_of_bookings', data = d1, order = d1.sort_values('num_of_bookings', ascending = False).agent)
```

##Chart-4

4. In which month most of the bookings happened?

```python
# groupby arrival_date_month and taking the hotel count
bookings_by_months_df=df.groupby(['arrival_date_month'])['hotel'].count().reset_index().rename(columns={'hotel':"Counts"})
# Create list of months in order
months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
# creating df which will map the order of above months list without changing its values.
bookings_by_months_df['arrival_date_month']=pd.Categorical(bookings_by_months_df['arrival_date_month'],categories=months,ordered=True)
# sorting by arrival_date_month
bookings_by_months_df=bookings_by_months_df.sort_values('arrival_date_month')

bookings_by_months_df
```

```python
# visualization code

# set plot size
plt.figure(figsize=(15,8))

#pltting lineplot on x- months & y- booking counts
sns.lineplot(x=bookings_by_months_df['arrival_date_month'],y=bookings_by_months_df['Counts'])

# set title for the plot
plt.title('Number of bookings across each month')
#set x label
plt.xlabel('Month')
#set y label
plt.ylabel('Number of bookings')
```
